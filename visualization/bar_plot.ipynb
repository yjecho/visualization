```python
# 데이터 전처리
df_hospitals = hospitals.loc[:, ["자치구", "계", "계.1"]].drop([0,1], axis = 0)

# 두 개의 axis를 가진 그래프 그리기
gu = np.array(df_hospitals.loc[:, "자치구"])
num_hospitals = np.array(df_hospitals.loc[:, "계"])
num_beds = np.array(df_hospitals.loc[:, "계.1"])

fig = plt.figure(figsize=(20,8)) ## Figure 생성 
fig.set_facecolor('white') ## Figure 배경색 지정
ax1 = fig.add_subplot() ## axes 생성
 
colors = sns.color_palette('summer', len(gu)) ## 바 차트 색상
 
xtick_label_position = list(range(len(gu))) ## x축 눈금 라벨이 표시될 x좌표
ax1.set_xticks(xtick_label_position) ## x축 눈금 
ax1.set_xticklabels(gu) ## x축 눈금 라벨
ax1.bar(xtick_label_position, num_hospitals, color=colors) ## 바차트 출력
 
color = 'blue'
ax2 = ax1.twinx() ## 새로운 axis 생성
ax2.plot(xtick_label_position, num_beds, color=color, linestyle='--', marker='o') ## 선 그래프 
ax2.tick_params(axis='y', labelcolor=color) ## 눈금 라벨 색상 지정

plt.show()
```
